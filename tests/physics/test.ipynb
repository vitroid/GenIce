```python
import matplotlib.pyplot as plt
import glob
import numpy as np


def load(filename):
    actual = dict()
    ideal = dict()
    Ndata = dict()
    with open(filename) as f:
        for line in f:
            cols = line.split()
            if len(cols) < 6:
                continue
            index = int(cols[0]), int(cols[1])
            ideal[index] = float(cols[4])
            actual[index] = float(cols[6])
            Ndata[index] = int(cols[5].split("/")[1])
    return ideal, actual, Ndata


# 
# https://bellcurve.jp/statistics/course/9490.html
def zvalue(p_actual, N, p_ideal):
    return (p_actual * N - p_ideal * N) / (p_ideal * N * (1 - p_ideal)) ** 0.5


# taken from ice with 2 million molecules
p_real = {
    (6, 0): 0.17466,
    (6, 1): 0.26519,
    (6, 3): 0.26244,
    (6, 5): 0.06351,
    (6, 7): 0.12542,
    (6, 9): 0.03659,
    (6, 11): 0.06984,
    (6, 21): 0.00236,
}


for typ in (0, 1, 3, 5, 7, 9, 11, 21):
    for s in (32, 16, 8, 4, 2):
        val = []
        out_of_confidence = 0
        for file in glob.glob(f"stats/{s}_*.ringstat"):
            ideal, actual, Ndata = load(file)
            if len(ideal) != 8:
                continue
            val.append(actual[6, typ])
            z = zvalue(actual[6, typ], Ndata[6, typ], p_real[6, typ])
            if not (-1.96 < abs(z) < 1.96):
                out_of_confidence += 1
        plt.hist(val, label=f"Size {s}x{s}x{s}", alpha=0.5, density=True)
        print(f"type {typ} size {s} samples {len(val)} average {np.mean(val)}")
        print(f"  Out of 95% confidence range {out_of_confidence} / {len(val)}")

    plt.stem(ideal[6, typ], 1, label="Bethe tree")
    plt.legend()
    plt.title(f"Type {typ}")
    plt.yscale("log")
    plt.show()
```
